#### Создать ноутбук, в котором выполнить решение вычислительной задачи (например, задачи из области физики, экономики, математики, статистики и т. д.) требующей построения трехмерного графика, условие которой предварительно необходимо согласовать с преподавателем.

имеем поверхность, заданную в векторной параметрической форме:


$\vec{r}(u,v) =
  \begin{cases}
    u*cos(v)       \\
    u*sin(v)  \\
    u^2
  \end{cases}$
  
построить график этой поверхности на интервале 
$u\in[0,2]$ и $v\in[0,2 * pi]$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

зададим значения $u$ и $v$ на интервалах, заданных в условии

```python
u = np.linspace(0, 2, 100)
v = np.linspace(0, 2*np.pi, 100)
u, v = np.meshgrid(u, v)
```

вычислим значения функции для каждой пары (u,v)


```python
x = u * np.cos(v)
y = u * np.sin(v)
z = u**2
```

создадим график поверхности


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

Данный трехмерный график показывает трехмерную поверхность, соответствующую функции $\vec{r}(u,v)$
